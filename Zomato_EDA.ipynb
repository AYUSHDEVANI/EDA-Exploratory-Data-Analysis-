```python
# Load the country code data
df_country = pd.read_excel("Zomatodataset/Country-Code.xlsx")
df_country.head()
```

```python
df_country.columns
```

```python
df.columns
```

```python
# Merge the both data set contry-code and zomato
final_df = pd.merge(df,df_country, on = "Country Code", how = "left")
```

```python
final_df.head()
```

```python
#  To check data types
final_df.dtypes
```

```python
final_df.columns
```

```python
final_df.Country.value_counts()
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

```python
# Plot Pie Chart - top 3 countries that use Zomato
plt.pie(x = country_val[:3],
        autopct = "%1.2f%%",
       labels = country_names[:3])
```

Observation: Zomato maximum records or transaction are from India After that USA and then United Kingdoms

```python
final_df.columns
```

```python
ratings = final_df.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index().rename(columns = {0:"Rating count"})
```

```python
ratings
```

## Obeservation
1. When Rating is between 4.5 to 4.9 --> Excellent
2. When RAting is between 4.0 to 4.4 --> Very Good
3. When Rating is between 3.5 to 3.9 --> Good
4. When Rating is between 2.5 to 3.4 --> Average
5. When Rating is between 1.8 to 2.4 --> Poor

```python
ratings.head()
```

```python

sns.barplot(data = ratings,
           x = "Aggregate rating",
           y = "Rating count")
```

```python
sns.barplot(data = ratings,
           x = "Aggregate rating",
           y = "Rating count",
           hue = "Rating color",
           palette = ["blue", "red", "orange", "yellow", "green", "darkgreen"])
```

Observation :
   1. Not Rated count is very high
   2. Maximum number of rating are between 2.5 to 3.4

```python
# Count plot
sns.countplot(x = "Rating color",
             data = ratings,
             palette = ["blue", "red", "orange", "yellow", "green", "darkgreen"])
```

```python
## Find out which currency is used by which Country?
final_df[["Country","Currency"]].groupby(["Country","Currency"]).size().reset_index().drop(columns = 0)
```
